```python
import logging

logging.basicConfig(
    filename="remove_duplicates.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def remove_duplicates(li: list) -> list:
    """
    Remove duplicate elements from list while preserving order.
    """
    logging.info("Removing duplicates while preserving order")

    result = []

    for item in li:
        if item not in result:
            logging.info(f"Adding unique element: {item}")
            result.append(item)
        else:
            logging.info(f"Duplicate found and skipped: {item}")

    logging.info(f"Final list after removing duplicates: {result}")
    return result


li = eval(input("Enter a list of integers: "))

result = remove_duplicates(li)

print("List after removing duplicates:", result)
```
