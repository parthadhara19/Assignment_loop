```python
import logging

logging.basicConfig(
    filename="check_sorted.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def is_sorted(li: list) -> bool:
    """
    Check whether the list is sorted in ascending order.
    """
    logging.info("Checking if list is sorted in ascending order")

    if len(li) < 2:
        logging.info("List has less than 2 elements, considered sorted")
        return True

    for i in range(len(li) - 1):
        if li[i] > li[i + 1]:
            logging.info("List is NOT sorted")
            return False

    logging.info("List is sorted in ascending order")
    return True


li = eval(input("Enter a list of integer: "))

result = is_sorted(li)

if result:
    print("List is sorted in ascending order")
else:
    print("List is NOT sorted")
```
