```python
import logging

logging.basicConfig(
    filename="reverse_string.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def reverse_string(s: str) -> str:
    """
    Reverse a given string using for loop (no slicing).
    """
    logging.info("Reversing the given string")

    rev = ""

    for ch in s:
        logging.info(f"Adding character {ch} to front")
        rev = ch + rev

    logging.info(f"Reversed string is: {rev}")
    return rev


s = input("Enter a string: ")

result = reverse_string(s)

print("Reversed string:", result)
```
