```python
cor_username = "Partha"
cor_pass = "1234"
attempts = 0

while attempts < 3:
    username = input("Enter username: ")
    password = input("Enter password: ")

    if username == cor_username and password == cor_pass:
        print("Login Successful")
        break
    else:
        attempts += 1
        print("Login unsuccessful")

if attempts == 3:
    print("Account locked due to 3 incorrect attempts.")
```
