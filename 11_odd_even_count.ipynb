```python
num = int(input("Enter a number: "))

def main(num):
    even = 0
    odd = 0

    while num > 0:
        digit = num % 10

        if digit % 2 == 0:
            even += 1
        else:
            odd += 1

        num = num // 10

    return even, odd

e, o = main(num)
print("Even digits:", e)
print("Odd digits:", o)

```
