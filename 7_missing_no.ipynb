```python
import logging

logging.basicConfig(
    filename="find_missing.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

def find_missing(li: list) -> int:
    """
    Find the missing number in a list containing numbers from 1 to N.
    """

    logging.info("Finding missing number in the list")

    n = len(li) + 1  
    expected_sum = n * (n + 1) // 2
    actual_sum = sum(li)

    logging.info(f"Expected sum from 1 to {n} is {expected_sum}")
    logging.info(f"Actual sum of list is {actual_sum}")

    missing = expected_sum - actual_sum

    logging.info(f"Missing number is {missing}")
    return missing



li = eval(input("Enter a list of integers from 1 to N (one missing): "))

result = find_missing(li)

print("Missing number:", result)
```
